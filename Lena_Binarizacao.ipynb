```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO

# Funo para carregar imagem da URL
def carregar_imagem_url(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    return cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)

# URL da imagem corrigida
url = 'https://raw.githubusercontent.com/sebavassou/binarizacaodeimagem/main/LENA%20colorida.jpg'

try:
    # Carregar a imagem
    imagem = carregar_imagem_url(url)

    # Converter para escala de cinza
    cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

    # Binarizao usando threshold
    _, binaria = cv2.threshold(cinza, 127, 255, cv2.THRESH_BINARY)

    # Plotar as trs imagens
    plt.figure(figsize=(15,5))

    plt.subplot(131)
    plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
    plt.title('Imagem Original')
    plt.axis('off')

    plt.subplot(132)
    plt.imshow(cinza, cmap='gray')
    plt.title('Escala de Cinza')
    plt.axis('off')

    plt.subplot(133)
    plt.imshow(binaria, cmap='gray')
    plt.title('Imagem Binarizada')
    plt.axis('off')

    plt.show()

except Exception as e:
    print(f"Erro ao carregar a imagem: {e}")
```
